# EduAI Career Advisor - Model Training Notebook
This notebook trains the Job Role Prediction model (`job_model.pkl`) and TF-IDF Vectorizer (`vectorizer.pkl`) used by the EduAI backend.

```python
import os
import pickle
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline

# Sample Training Dataset for Career Job Role Classification
data = [
    # Data Science
    ("python machine learning deep learning tensorflow pytorch scikit-learn nlp data science pandas numpy statistics", "Data Science"),
    ("data analysis pandas numpy power bi tableau sql statistics predictive modeling data analyst", "Data Science"),
    ("computer vision neural networks xgboost random forest clustering regression python scikit-learn", "Data Science"),
    
    # Python Developer
    ("python django flask fastapi rest api postgresql web development python developer celery redis", "Python Developer"),
    ("python backend developer microservices docker unit testing git fastapi sqlalchemy", "Python Developer"),
    
    # Web Designing / Frontend / Fullstack
    ("html css javascript react vue angular next.js node.js bootstrap tailwind frontend web designer fullstack", "Web Designing"),
    ("javascript react redux typescript html5 css3 responsive design web development ui ux", "Web Designing"),
    
    # Java Developer
    ("java spring boot hibernate maven j2ee jsp microservices mysql RESTful APIs", "Java Developer"),
    ("java core java spring framework backend development junit gradle postgresql", "Java Developer"),
    
    # DevOps Engineer
    ("aws docker kubernetes jenkins terraform ansible ci/cd devops linux bash cloud azure gcp", "DevOps Engineer"),
    ("cloud engineering devops site reliability aws lambda docker kubernetes monitoring prometheus grafana", "DevOps Engineer"),
    
    # Database
    ("mysql postgresql mongodb oracle sql server nosql pl/sql database snowflake redis dba", "Database"),
    ("database administration sql query optimization indexing backup recovery mongodb postgresql", "Database"),
    
    # DotNet Developer
    ("c# .net asp.net dotnet mvc web api entity framework sql server visual studio", "DotNet Developer"),
    
    # Blockchain
    ("solidity smart contracts ethereum web3 blockchain crypto dapps hardhat truffle", "Blockchain"),
    
    # Testing / QA
    ("selenium testing qa cucumber test automation manual testing jira testng cypress bug tracking", "Testing"),
    
    # Business Analyst
    ("business analyst agile scrum requirements gathering stakeholder management visio bpms user stories", "Business Analyst"),
    
    # HR
    ("recruitment onboarding payroll hrms employee relations talent acquisition hr sourcing interview", "HR"),
    
    # Sales & Marketing
    ("b2b lead generation crm salesforce negotiation digital marketing seo marketing google ads", "Sales")
]

df = pd.DataFrame(data, columns=["skills", "job_role"])
print(f"Dataset loaded with {len(df)} samples.")
df.head()
```

```python
# Train Vectorizer & Model
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(df["skills"])
y = df["job_role"]

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
print("Model trained successfully!")
```

```python
# Save model and vectorizer to backend/models/
out_dir = os.path.join("backend", "models")
os.makedirs(out_dir, exist_ok=True)

with open(os.path.join(out_dir, "job_model.pkl"), "wb") as f:
    pickle.dump(model, f)

with open(os.path.join(out_dir, "vectorizer.pkl"), "wb") as f:
    pickle.dump(vectorizer, f)

print(f"Saved job_model.pkl and vectorizer.pkl to {out_dir}/")
```
